## Normal (Gaussian) Distribution

```python
n = np.arange(-50, 50)
mean = 0
normal = stats.norm.pdf(n, mean, 10)
plt.plot(n, normal)
plt.xlabel('Distribution', fontsize=12)
plt.ylabel('Probability', fontsize=12)
plt.title("Normal Distribution")
```

## Uniform Distribution

```python
import warnings
warnings.simplefilter("ignore", UserWarning)

# Create an array of random numbers from uniform distribution
uniform = np.random.uniform(-4,4,1000000)

# Let's plot them
ax = sns.distplot(uniform, label='Uniform Distribution')
legend = plt.legend()
plt.show()
```

## Log Normal Distribution

```python
# using np.lognormal() method
data = np.random.lognormal(0.5, 3, 20)
  
# Let's plot them
ax = sns.distplot(data, label='Log Normal Distribution')
legend = plt.legend()
plt.show()
```

## Poisson Distribution

```python
# n = number of events, lambd = expected number of events 
# which can take place in a period
for lambd in range(2, 8, 2):
    n = np.arange(0, 10)
    poisson = stats.poisson.pmf(n, lambd)
    plt.plot(n, poisson, '-o', label="λ = {:f}".format(lambd))
    plt.xlabel('Number of Events', fontsize=12)
    plt.ylabel('Probability', fontsize=12)
    plt.title("Poisson Distribution varying λ")
    plt.legend()
```
